```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
# --- Data ---
# 0: cat, 1: dog, 2: squirrel
predictions = [0, 0, 1, 1, 0, 0, 2, 2, 1, 0]
ground_truth = [0, 1, 1, 1, 1, 0, 2, 2, 2, 2]
class_names = ['cat', 'dog', 'squirrel']
```

```python
# --- Calculations ---

# 1. Generate the confusion matrix
cm = confusion_matrix(ground_truth, predictions)

# 2. Generate accuracy
accuracy = accuracy_score(ground_truth, predictions)

# 3. Generate classification report
report = classification_report(ground_truth, predictions, target_names=class_names)
```

```python
# --- Output Results ---
print("--- Model Evaluation Matrix ---")
print(f"Accuracy: {accuracy:.2f}\n")
print("Classification Report:")
print(report)
print("Confusion Matrix:")
print(cm)
```

```python
# --- Plotting the confusion matrix ---
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix')
plt.show()
```
